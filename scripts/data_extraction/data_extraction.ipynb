```python
import yfinance as yf
import pandas as pd
```

```python
ticker = "^GSPC"

data = yf.download(ticker, start="2015-01-01", end="2025-01-01")

data = data.reset_index()

data.to_csv("../../data/raw/sp500_index.csv", index=False)

print("Market index data downloaded successfully.")
```

```python
import yfinance as yf
import pandas as pd

tickers = ["AAPL", "MSFT", "AMZN", "NVDA", "TSLA"]

all_data = []

for ticker in tickers:
    df = yf.download(ticker, start="2015-01-01", end="2025-01-01")
    df["Ticker"] = ticker
    df = df.reset_index()
    all_data.append(df)

stock_data = pd.concat(all_data)

stock_data.to_csv("../../data/raw/stock_prices.csv", index=False)

print("Stock data extracted successfully.")
```
